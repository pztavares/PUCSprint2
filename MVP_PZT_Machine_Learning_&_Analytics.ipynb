## Contextualização do Problema

As **estatísticas públicas** são dados e informações oficiais que o Estado produz e divulga para entender a realidade social, econômica e ambiental de um país, sendo cruciais para subsidiar políticas públicas, planejamento e o debate público (IBGE, 2025).

As principais fontes para esses dados são os sistemas estatísticos nacionais de cada país, **divulgados em sites ou portais de dados oficiais governamentais** (ODIN, 2025). O IBGE (Instituto Brasileiro de Geografia e Estatística) é o principal órgão responsável pela produção de estatísticas oficiais no Brasil (IBGE, 2025).

## Levantamento dos Dados

Para o levantamento dos dados desse trabalho, a pesquisadora acessou o **Open Data Inventory (ODIN)**, que oferece desde 2016 uma avaliação da **cobertura e abertura dos dados** disponibilizados nos sites governamentais de 197 países. As pontuações seguem uma escala de 0 a 10, onde 0 representa nota mínima e 10 representa a máxima (ODIN, 2025).

O ODIN mede a integralidade da oferta estatística de um país (chamada de **"cobertura"**) e o grau de atendimento desses
conjuntos de dados aos padrões internacionais de abertura de dados (chamada de **"abertura"**) (ODIN, 2025; OKF, 2025).

Em resumo, a pontuação geral do ODIN é um **indicador de quão completas e abertas são as ofertas de dados do sistema estatístico de cada país** (ODIN, 2025).

### Objetivo do estudo

O objetivo principal é treinar um modelo de classificação em Machine Learning e classificar os dados estatísticos de um país em três categorias (variável "target"), com base em dez características ligadas à cobertura e abertura de dados (descritas abaixo).

## Definição de cobertura de dados
A cobertura dos dados é medida por cinco elementos que medem a abrangência da oferta de dados do país (ODIN, 2025):
1. Disponibilidade de indicadores e desagregações
2. Disponibilidade de dados nos últimos cinco anos
3. Disponibilidade de dados nos últimos dez anos
4. Disponibilidade de dados no primeiro nível geográfico subnacional
5. Disponibilidade de dados no segundo nível geográfico subnacional

## Definição de abertura de dados

Qualquer pessoa deve ter a liberdade para acessar, usar e compartilhar os dados desses sites, ou seja, **os dados devem ser "abertos"**. A abertura dos dados é medida por cinco elementos que avaliam o quanto as ofertas de dados de um país atendem aos padrões internacionais de abertura (OKF, 2025):
1. Disponibilidade de dados em formato legível por máquina
2. Disponibilidade de dados em formato não proprietário
3. Disponibilidade de metadados de referência
4. Disponibilidade de opções de download que tornam os dados mais acessíveis: downloads em massa, APIs, opções de exportação de dados personalizáveis
5. Disponibilidade de uma licença de dados abertos.

### Tipo de problema

Este é um problema de **classificação supervisionada**. Dado um conjunto de dez características (atributos de cobertura e abertura de dados), o objetivo é prever a qual das três classes (categoria estatística) um conjunto de dados estatístico pertence.

## Modelagem e Inferência

### Criação e avaliação de modelos: linha base

Para avaliar o desempenho de diferentes modelos, inicialmente treinaremos 11 algoritmos distintos e, em seguida, selecionaremos os que apresentarem melhor desempenho para uma etapa posterior de otimização.

```python
np.random.seed(SEED)

# Lista que armazenará os modelos
models = []

# Criando os modelos e adicionando-os na lista de modelos
models.append(('LR', LogisticRegression(max_iter=200)))
models.append(('KNN', KNeighborsClassifier()))
models.append(('CART', DecisionTreeClassifier()))
models.append(('NB', GaussianNB()))
models.append(('SVM', SVC()))

# Definindo os parâmetros do classificador base para o BaggingClassifier
base = DecisionTreeClassifier()
num_trees = NUM_TREES
max_features = MAX_FEATURES

# Criando os modelos para o VotingClassifier
bases = []
model1 = LogisticRegression(max_iter=200)
bases.append(('logistic', model1))
model2 = DecisionTreeClassifier()
bases.append(('cart', model2))
model3 = SVC()
bases.append(('svm', model3))

# Criando os ensembles e adicionando-os na lista de modelos
models.append(('Bagging', BaggingClassifier(estimator=base, n_estimators=num_trees)))
models.append(('RF', RandomForestClassifier(n_estimators=num_trees, max_features=max_features)))
models.append(('ET', ExtraTreesClassifier(n_estimators=num_trees, max_features=max_features)))
models.append(('Ada', AdaBoostClassifier(n_estimators=num_trees)))
models.append(('GB', GradientBoostingClassifier(n_estimators=num_trees)))
models.append(('Voting', VotingClassifier(bases)))

# Listas para armazenar os resultados
results = []
names = []

# Avaliação dos modelos
for name, model in models:
    cv_results = cross_val_score(model, X_train, y_train, cv=kfold, scoring=scoring)
    results.append(cv_results)
    names.append(name)
    msg = "%s: %f (%f)" % (name, cv_results.mean(), cv_results.std())
    print(msg)

# Boxplot de comparação dos modelos
fig = plt.figure(figsize=(15,10))
fig.suptitle('Gráfico 5 - Comparação dos Modelos')
ax = fig.add_subplot(111)
plt.boxplot(results)
ax.set_xticklabels(names)
plt.show()
```

No gráfico 5, pode-se notar os 5 modelos com melhor desempenho para este conjunto de dados: Random Forest (RF), Bagging, Gradient Boosting (GB), ExtraTrees (ET) e Voting. A próxima etapa é o processo de normalização e padronização, visando melhorar os resultados.


### Criação e avaliação de modelos: dados padronizados e normalizados

```python
np.random.seed(SEED)

# Listas para armazenar os armazenar os pipelines e os resultados para todas as visões do dataset
pipelines = []
results = []
names = []

# Criando os elementos do pipeline

# Algoritmos que serão utilizados
random_forest = ('RF', RandomForestClassifier(n_estimators=num_trees, max_features=max_features))
bagging = ('Bag', BaggingClassifier(estimator=base, n_estimators=num_trees))
gradient_boosting = ('GB', GradientBoostingClassifier(n_estimators=num_trees))
extra_trees = ('ET', ExtraTreesClassifier(n_estimators=num_trees, max_features=max_features))
voting = ('Voting', VotingClassifier(bases))

# Transformações que serão utilizadas
standard_scaler = ('StandardScaler', StandardScaler())
min_max_scaler = ('MinMaxScaler', MinMaxScaler())

# Montando os pipelines

# Dataset original
pipelines.append(('RF-orig', Pipeline([random_forest])))
pipelines.append(('Bag-orig', Pipeline([bagging])))
pipelines.append(('GB-orig', Pipeline([gradient_boosting])))
pipelines.append(('ET-orig', Pipeline([extra_trees])))
pipelines.append(('Vot-orig', Pipeline([voting])))

# Dataset Padronizado
pipelines.append(('RF-padr', Pipeline([standard_scaler, random_forest])))
pipelines.append(('Bag-padr', Pipeline([standard_scaler, bagging])))
pipelines.append(('GB-padr', Pipeline([standard_scaler, gradient_boosting])))
pipelines.append(('ET-padr', Pipeline([standard_scaler, extra_trees])))
pipelines.append(('Vot-padr', Pipeline([standard_scaler, voting])))

# Dataset Normalizado
pipelines.append(('RF-norm', Pipeline([min_max_scaler, random_forest])))
pipelines.append(('Bag-norm', Pipeline([min_max_scaler, bagging])))
pipelines.append(('GB-norm', Pipeline([min_max_scaler, gradient_boosting])))
pipelines.append(('ET-norm', Pipeline([min_max_scaler, extra_trees])))
pipelines.append(('Vot-norm', Pipeline([min_max_scaler, voting])))

# Executando os pipelines
for name, model in pipelines:
    cv_results = cross_val_score(model, X_train, y_train, cv=kfold, scoring=scoring)
    results.append(cv_results)
    names.append(name)
    msg = "%s: %.3f (%.3f)" % (name, cv_results.mean(), cv_results.std()) # formatando para 3 casas decimais
    print(msg)

# Boxplot de comparação dos modelos
fig = plt.figure(figsize=(25,6))
fig.suptitle('Gráfico 6 - Comparação dos Modelos - Dataset orginal, padronizado e normalizado')
ax = fig.add_subplot(111)
plt.boxplot(results)
ax.set_xticklabels(names, rotation=90)
plt.show()
```

* Após a etapa dos pipelines, os três melhores modelos foram Random Forest (RF), Bagging e Gradient Boosting (GB). Houve pouca variação nos resultados entre os originais e os dados padronizados e normalizados.

* O Random Forest (RF) manteve a superioridade na precisão e métricas, então, a etapa seguinte será encontrar a melhor combinação de hiperparâmetros para maximizar o desempenho do modelo.

### Otimização dos hiperparâmetros

```python
# Define the Random Forest Classifier
rf_classifier = RandomForestClassifier(n_estimators=num_trees, max_features=max_features)

# Define the parameter grid for tuning
param_grid = {
    'n_estimators': [1,3,5],
    'max_depth': [3, 4, 5],
}

# Perform GridSearchCV
grid_search = GridSearchCV(estimator=rf_classifier, param_grid=param_grid,
                           cv=kfold, scoring=scoring, n_jobs=-1, verbose=1)

grid_search.fit(X_train, y_train)

# Print the best parameters and best score
print(f"Best parameters: {grid_search.best_params_}")
print(f"Best score: {grid_search.best_score_}")

# Evaluate the best model on the test set
best_gb_model = grid_search.best_estimator_
test_accuracy = best_gb_model.score(X_test, y_test)
print(f"Test accuracy with best parameters: {test_accuracy}")
```
