### Contextualização do Problema
--------------------------------
A **igualdade de gênero** é um assunto que vem sido muito discutido ultimamente e é o objetivo de número 5 (ODS 5) dos Objetivos do Desenvolvimento Sustentável (ODS), propostos pela Organização das Nações Unidas (ONU). O ODS 5 propõe-se a “Alcançar a igualdade de gênero e empoderar todas as mulheres e meninas”.

Apesar dos avanços das mulheres na reinvindicação de seus direitos, **ainda existem muitas barreiras** a serem transpostas para garantir que tenham as mesmas oportunidades e tratamento, independente do gênero. Isso envolve combater a discriminação, a violência e a disparidade salarial, além de promover a equidade na participação política, econômica e social.

A ação governamental, através das **estatísticas públicas**, vem apresentando **dados abertos** em prol da paridade de gênero, visando ampliar as políticas e estratégias para melhorar a representação das mulheres na força de trabalho e em cargos de liderança, bem como a igualdade salarial.

Na **área educacional**, destaca-se o investimento no conhecimento das meninas e mulheres, para que possam realizar seu potencial como membros produtivos da sociedade. A educação é uma ferramenta-chave para o empoderamento feminino, a conscientização sobre direitos e para a igualdade de oportunidades na vida profissional e social.

O problema é sentido em todas as áreas profissionais, mas ultimamente tem sido **fortemente citado nas áreas STEM**, que á sigla em inglês usada para _Science, Technology, Engineering, and Mathematics_ (em português, Ciências, Tecnologia, Engenharia e Matemática).

### Estrutura do Esquema Floco de Neve (_"snowflake"_)
-------------
Para organizar os dados, utilizou-se o esquema em floco de neve (*snowflake*), que permite um design normalizado, evitando a duplicação de dados. 

No modelo, a tabela fato (em azul) reúne os percentuais referentes ao ensino superior mundial, classificados por país e por ano. Essa tabela ocupa a posição central do modelo, sendo circundada pelas tabelas de dimensão (em rosa), estruturadas de forma normalizada.

**Tabela Fato:** `fato_graduacao`

A tabela central do esquema apresenta, para cada país, o número de mulheres e homens graduados anualmente no ensino superior, expresso em percentagem na área de estudo em questão.

**Tabelas Dimensão:** `dim_pais, dim_nivel_renda, dim_regiao_geo, dim_idh, dim_campo_estudo`.

As tabelas de dimensão reúnem os atributos que descrevem a tabela fato. Elas apresentam detalhes sobre as áreas de estudo e informações sobre os países, incluindo sua região geográfica e sua posição no cenário mundial, considerando indicadores como IDH e o nível de renda. Cada dimensão possui uma coluna de chave primária e é vinculada à tabela fato por meio de relacionamentos de chave estrangeira.

```python
%sql

USE mvp.graduacao;
SELECT *
FROM  (SELECT COUNT(*) AS Graduacao FROM fato_graduacao),
      (SELECT COUNT(*) AS CampoEstudo FROM dim_campo_estudo),
      (SELECT COUNT(*) AS IDH FROM dim_idh),
      (SELECT COUNT(*) AS NivelRenda FROM dim_nivel_renda),
      (SELECT COUNT(*) AS Pais FROM dim_pais),
      (SELECT COUNT(*) AS Regiao FROM dim_regiao_geo)
```


**Quantidade de registros das tabelas:**
- `fato_graduacao`: 11.901 registros

- `dim_nivel_renda`: 4 registros

**1) Quais são as áreas de estudo e concentração de estudantes do sexo masculino e feminino em todo o mundo?**

**2) A disparidade de gênero em STEM é um padrão mundial ou pode estar concentrada por região geográfica?**

```python
%sql
USE mvp.graduacao;
SELECT R.DESC_REG_GEOGR_POR AS Regiao, AVG(G.PERC_MULHERES/100) AS Mulheres, AVG(G.PERC_HOMENS/100) AS Homens 
FROM fato_graduacao G, dim_pais P, dim_regiao_geo R
WHERE G.ID_PAIS = P.ID_PAIS
AND P.ID_REG_GEOGR = R.ID_REG_GEOGR
AND G.ID_CAMPO_ESTUDO = 7
GROUP BY R.DESC_REG_GEOGR_POR
ORDER BY Mulheres DESC
```

```python

# Converter _sqldf para Pandas DataFrame
df_pandasq2 = _sqldf.toPandas()

df_pandasq2['Mulheres'] = pd.to_numeric(df_pandasq2['Mulheres'], errors='coerce')
df_pandasq2['Homens'] = pd.to_numeric(df_pandasq2['Homens'], errors='coerce')

df_pandasq2 = df_pandasq2.sort_values(by='Mulheres', ascending=True)

df_pandasq2 = df_pandasq2.dropna(subset=['Mulheres', 'Homens'])

# Create the stacked bar chart
ax = df_pandasq2.plot.barh(
    x='Regiao',
    y=['Mulheres', 'Homens'],
    stacked=True,
    title='Distribuição por Gênero',
    color=['pink', 'skyblue']
)

plt.bar_label(ax.containers[0], label_type='center', fmt='%.0f')
plt.bar_label(ax.containers[1], label_type='center', fmt='%.0f')

plt.xlabel('Percentual')
plt.ylabel('Região Geográfica')
plt.title('Gráfico 2 - Percentual de mulheres e homens em STEM, por região geográfica')
plt.xticks(rotation=0)
plt.legend(title='Gênero')

plt.show()
```

A análise da distribuição por gênero em STEM nas diferentes regiões do mundo revela que os **homens continuam predominando, independentemente do contexto geográfico**. Embora a intensidade da desigualdade varie entre as regiões, o padrão de sub-representação feminina é universal. As maiores disparidades aparecem no Sul da Ásia e na África Subsaarieana, onde os homens representam cerca de 76% do total, deixando às mulheres uma participação pouco superior a 23%. Esse desequilíbrio indica barreiras estruturais mais acentuadas nessas regiões, que limitam o acesso e a permanência das mulheres em determinados campos profissionais ou educacionais.

Por outro lado, regiões como o Oriente Médio e Norte da África, assim como a Europa e a Ásia Central, apresentam percentuais relativamente mais equilibrados, com a participação feminina variando entre 34% e quase 40%. Ainda que a desigualdade persista, esses números sugerem maior avanço na inclusão das mulheres. Nas Américas, tanto a América Latina e Caribe quanto a América do Norte mostram um cenário intermediário, no qual os homens compõem aproximadamente 67% a 70% da distribuição, evidenciando que a sub-representação feminina permanece significativa, embora menos pronunciada que nas regiões mais desiguais.

De forma geral, os dados reforçam que a desigualdade de gênero é um fenômeno global e persistente, caracterizado por maior presença masculina em todos os contextos analisados. Essa tendência revela a necessidade contínua de políticas e iniciativas capazes de ampliar a participação das mulheres, especialmente nas regiões onde a disparidade ainda é mais acentuada.

Esse desequilíbrio sugere que, apesar de maior acesso geral à educação e ao mercado de trabalho, a presença feminina em determinadas áreas, especialmente as mais valorizadas economicamente, permanece limitada, possivelmente devido à persistência de barreiras estruturais e culturais.

Nos países de renda baixa, a diferença entre homens e mulheres é menos acentuada: 60,14% de homens e 39,86% de mulheres. Essa participação relativamente maior das mulheres pode estar relacionada à necessidade econômica que impulsiona uma inserção mais ampla no mercado, ou a padrões educacionais distintos, que nem sempre seguem as divisões tradicionais observadas em países mais desenvolvidos.

Isso indica que, mesmo em contextos com maior expansão educacional, a desigualdade de gênero ainda é significativa e não acompanha de forma proporcional o crescimento econômico.

De modo geral, apesar da expectativa de que países mais ricos apresentem maior equidade, os números mostram que **a desigualdade de gênero persiste mesmo em contextos de alto desenvolvimento**, reforçando a necessidade de políticas específicas para promover a participação equilibrada entre homens e mulheres.


**4) Em quais países as mulheres ultrapassam os homens em número de formadas em áreas STEM?**

```python
%sql
USE mvp.graduacao;
SELECT P.DESC_PAIS_POR AS Pais, G.ANO AS Ano, G.PERC_MULHERES/100 Mulheres, G.PERC_HOMENS/100 AS Homens 
FROM fato_graduacao G, dim_pais P
WHERE G.ID_PAIS = P.ID_PAIS
AND G.ID_CAMPO_ESTUDO = 7
AND G.PERC_MULHERES > G.PERC_HOMENS
ORDER BY Mulheres DESC, G.ANO
```

```python

# Converter _sqldf para Pandas DataFrame
df_pandasq4 = _sqldf.toPandas()

df_pandasq4['Mulheres'] = pd.to_numeric(df_pandasq4['Mulheres'], errors='coerce')
df_pandasq4['Homens'] = pd.to_numeric(df_pandasq4['Homens'], errors='coerce')

df_pandasq4 = df_pandasq4.head(10)
df_pandasq4 = df_pandasq4.sort_values(by=['Mulheres', 'Pais'], ascending=[True, False])

df_pandasq4 = df_pandasq4.dropna(subset=['Mulheres', 'Homens'])

# Create a new column for combined label
df_pandasq4['Pais_Ano'] = df_pandasq4['Pais'] + ' (' + df_pandasq4['Ano'].astype(str) + ')'

# Create the stacked bar chart
ax = df_pandasq4.plot.barh(
    x='Pais_Ano',
    y=['Mulheres', 'Homens'],
    stacked=True,
    title='Países com maior concentração de mulheres em STEM',
    color=['pink', 'skyblue']
)

plt.bar_label(ax.containers[0], label_type='center', fmt='%.0f')
plt.bar_label(ax.containers[1], label_type='center', fmt='%.0f')

plt.xlabel('Percentual')
plt.ylabel('Pais')
plt.title('Gráfico 4 - Dez países com maior concentração de mulheres em STEM, por ano')
plt.xticks(rotation=0)
plt.legend(title='Gênero')

plt.show()
```

Os dados revelam que, em certos anos, a proporção de mulheres superou a de homens em alguns países, com variações situadas entre 50% e 75%. 

**Antígua e Barbuda**, por exemplo, tem um cenário STEM em crescimento, com foco em engajar jovens através de programas como Olimpíadas, bolsas de estudo e participação em competições internacionais, mostrando um compromisso com a inovação e educação tecnológica no Caribe (Fonte: internet). 

A alta proporção de mulheres graduadas em STEM em **Brunei** vem do esforço do país para aumentar a participação feminina em STEM, com programas que abordam barreiras como lacunas de liderança e estereótipos (Fonte: Internet). 

Alguns **países africanos** aparecem na lista (Argélia, Benin, Eswatini, Libéria, Serra Leoa e Tunísia), pois o continente recebe investimentos internacionais em tecnologia e inovação para incentivar o crescimento económico e combater a pobreza. Várias dessas iniciativas têm como objetivo reduzir a lacuna de género e empoderar mulheres em áreas de tecnologia e STEM (Fonte: Internet). 

Esses casos atípicos indicam avanços contínuos na inclusão feminina em STEM:
- Maior acesso das mulheres ao ensino superior em determinados países.
- Políticas de incentivo.
- Diferenças culturais na escolha de carreiras.

Contudo, mesmo diante dessa superioridade feminina, **apenas 75 registros apresentam esse comportamento**, indicando tratar-se de situações pouco expressivas frente ao conjunto geral dos dados sobre STEM (1.223 registros).


**5) Países com Índice de Desenvolvimento Humano (IDH) alto tendem a possuir um Índice de Desigualdade de Gênero (IDG) mais baixo?**

```python
%sql
USE mvp.graduacao;
SELECT  P.DESC_PAIS_POR AS Pais,
        CAST(REPLACE(P.IDH_PAIS, ',', '.') AS DECIMAL(6, 3)) AS IDH, 
        CAST(REPLACE(P.IDG_PAIS, ',', '.') AS DECIMAL(6, 3)) AS IDG 
FROM dim_pais P
WHERE P.ID_IDH_PAIS = 1
AND P.IDH_PAIS IS NOT NULL
AND P.IDG_PAIS IS NOT NULL
ORDER BY P.IDH_PAIS DESC 
```

```python
# IDH alto - maior desenvolvimento humano
# IDG baixo - baixa desigualdade de gênero

df_pandasq5 = _sqldf.toPandas()

df_pandasq5.plot(kind='scatter', x='IDH', y='IDG', s=32, alpha=.8)
plt.gca().spines[['top', 'right',]].set_visible(False)
plt.title('Gráfico 5 - Correlação entre o IDH e o IDG')
```

Para essa análise, são apresentados os países classificados pelo Banco Mundial como "IDH muito alto". 

No gráfico 5, há uma tendência negativa clara entre o IDH e o IDG: quando o valor do IDH aumenta, o IDG tende a diminuir, ou seja, existe uma correlação negativa entre as duas variáveis.

Em outras palavras, os dados mostram que, em geral, países com maior IDH apresentam menor desigualdade de gênero, ou seja, com IDG muito próximo a zero. Nações como Noruega, Dinamarca, Islândia e Suíça combinam alto desenvolvimento humano com índices muito baixos de desigualdade, indicando que acesso à educação, saúde e renda contribui para maior igualdade entre homens e mulheres. 

Entretanto, alguns países com IDH elevado, como Estados Unidos, Japão e nações do Oriente Médio, ainda apresentam desigualdade significativa, mostrando que desenvolvimento econômico sozinho não garante igualdade de gênero, sendo também necessários fatores culturais, sociais e políticas públicas. 

Outros países, especialmente na América Latina e Europa Oriental, o IDG tende a ser maior, evidenciando que, mesmo com avanços no desenvolvimento, persistem diferenças entre homens e mulheres em oportunidades e direitos. 

Em resumo, **existe uma tendência geral de que quanto maior o IDH, menor o IDG**, embora essa relação não seja absoluta.


**6) Em um país com um Índice de Desigualdade de Gênero (IDG) baixo existe predominância de mulheres graduadas em STEM?**

Para responder a essa pergunta, vamos analisar dois países:
- Dinamarca
- Tunísia

No gráfico 6, a largura do violino na Dinamarca mostra que os percentuais concentram-se aproximadamente entre 28% e 38%, indicando uma distribuição relativamente compacta, com menor variação ao longo do tempo. Em contrapartida, na Tunísia, apesar de um contexto nacional marcado por desigualdade de gênero mais elevada (IDG = 0,238), as mulheres apresentam predominância no campo STEM, com percentuais entre 43% a 55%.

Esse fenômeno, chamado de _"The Gender-Equality Paradox in STEM Education"_, desafia a ideia de que a igualdade aumentaria a presença feminina em STEM. Pesquisadores que sugerem que, em sociedades menos igualitárias, a escolha por carreiras STEM pelas mulheres pode ocorrer por necessidade de independência econômica ou como estratégia de ascensão social e empoderamento. Em contextos mais igualitários, as mulheres têm maior liberdade para seguir seus interesses individuais.

**8) Como tem evoluído a participação de mulheres nas áreas STEM no mundo ao longo do tempo?**

```python
%sql
USE mvp.graduacao;
SELECT G.ANO AS Ano, AVG(G.PERC_MULHERES/100) AS Mulheres, AVG(G.PERC_HOMENS/100) AS Homens 
FROM fato_graduacao G
WHERE G.ID_CAMPO_ESTUDO = 7
GROUP BY G.ANO 
ORDER BY G.ANO
```

```python
# Converter _sqldf para Pandas DataFrame
df_pandasq8 = _sqldf.toPandas()

# Criando o gráfico
plt.figure(figsize=(12,6))
ax = sns.lineplot(data=df_pandasq8, x='Ano', y='Mulheres', marker="o", linewidth=2.5, color='b')

# Função para formatar os valores do eixo Y
def format_y_axis(x, pos):
    if x >= 1e6:
        return f'{x*1e-6:.1f}M'
    if x >= 1e3:
        return f'{x*1e-3:.0f}k'
    return f'{x:.0f}'

# Aplicar a formatação personalizada ao eixo Y
ax.yaxis.set_major_formatter(plt.FuncFormatter(format_y_axis))

plt.xlabel("Ano", fontsize=12)
plt.ylabel("% Mulheres", fontsize=12)
plt.title("Gráfico 8 - Percentual de graduação de mulheres no mundo, por ano", fontsize=14)
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

Os dados revelam uma **tendência consistente de crescimento da participação feminina nas áreas STEM** entre 1998 e 2019. Em 1998, as mulheres representavam aproximadamente 19,6% do total, enquanto em 2019 esse valor alcança cerca de 35,1%. Apesar das oscilações observadas em alguns anos (especialmente em 2008 e 2012) o avanço geral demonstra uma trajetória de expansão gradual.

A participação masculina, por sua vez, apresentou redução proporcional, passando de cerca de 80% em 1998 para aproximadamente 65% em 2019, o que indica uma diminuição significativa do desequilíbrio de gênero no setor. Mesmo assim, o percentual feminino mantém-se distante da paridade, sugerindo que persistem barreiras estruturais que afetam o ingresso e a permanência das mulheres em carreiras científicas e tecnológicas.

Em síntese, o período de 1998 a 2019 caracteriza-se por um crescimento expressivo, porém moderado, da presença feminina em STEM, **evidenciando avanços importantes, ainda que insuficientes**, em direção a uma maior equidade de gênero nessas áreas.


## Conclusões

Os dados mostram que a distribuição de homens e mulheres por área de estudo revela **forte segregação de gênero**. Campos ligados ao cuidado e às ciências humanas permanecem femininos, enquanto tecnologia, engenharia e setores industriais seguem dominados por homens. Algumas áreas mostram equilíbrio, mas ainda são exceção (Gráfico 1).
